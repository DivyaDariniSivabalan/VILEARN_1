```python
x = df['Gender'].value_counts()
```

```python
sns.violinplot(x='Gender',y='Age',data=df,palette=['pink','green'])
```

```python
plt.pie(x,labels=x.index,autopct='%1.1f%%')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['day'] = df['Date'].dt.day
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
```

```python
df.head(5)
```

```python
plt.pie(df['month'].value_counts(),labels=df['month'].value_counts().index,autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
sns.lineplot(x="year",y="Total Amount",data=df)
```

```python
sns.pairplot(df, hue="Product Category")
```

```python
df["Product Category"].value_counts()
```

```python
sns.countplot(x=df["Product Category"])
```

```python
p = df.groupby("Product Category")[["Quantity","Total Amount"]].sum()
p
```

```python
p.plot(kind="bar", figsize=(12,5))
plt.title("Quantity v/s Total Amount")
plt.ylabel("Quantity and Total Amount")
plt.show()
```
